```python
import pandas as pd

df = pd.read_csv('Heart_disease_statlog.csv')

print(df.head())

```

```python
age = 'age'

maximum = df[age].max()
minimum = df[age].min()
average = df[age].mean()

print(maximum)
print(minimum)
print(average)
```

```python
percent_diff = (maximum - average) / average * 100
print(percent_diff)

```

```python
df[age + '_normalized'] = (df[age] - minimum) / (maximum - minimum)
print(df.head())
```

```python
korelacija = df.corr()

korelacija_unstacked = korelacija.unstack()
sorted_correlations = korelacija_unstacked.sort_values(ascending=False)

biggest_positive = sorted_correlations[sorted_correlations < 1].idxmax()
print(biggest_positive)
print(sorted_correlations[biggest_positive])

biggest_negative = sorted_correlations.idxmin()
print(biggest_negative)
print(sorted_correlations[biggest_negative])
```

```python
from selenium import webdriver
import time
import re
from bs4 import BeautifulSoup

driver = webdriver.Firefox()
time.sleep(5)
driver.get("https://www.realitica.com/")
time.sleep(5)


with open("realitica_links.txt", "r", encoding="utf-8") as f:
    listing_urls = [line.strip() for line in f.readlines()]

def extract_field(soup, field_name, regex=False, as_int=False, as_float=False, as_bool=False):
    try:
        if regex:
            tag = soup.find(text=re.compile(field_name))
        else:
            tag = soup.find("strong", string=field_name)
        if not tag:
            return "Nema"
        val = tag.next_sibling
        if val and isinstance(val, str):
            val = val.strip().replace("€", "").replace(",", "")
        if as_int:
            return int(re.findall(r'\d+', val)[0]) if val else 0
        elif as_float:
            return float(re.findall(r'\d+', val)[0]) if val else 0.0
        elif as_bool:
            return True
        return val
    except:
        return "Nema"

output_data = []

for url in listing_urls:
    js_script = f"""
    return fetch("{url}")
        .then(res => res.text())
        .then(html => html);
    """
    try:
        html = driver.execute_script(js_script)
        time.sleep(1)
        soup = BeautifulSoup(html, "html.parser")

        data = {
            "vrsta": extract_field(soup, "Vrsta"),
            "područje": extract_field(soup, "Područje"),
            "lokacija": extract_field(soup, "Lokacija"),
            "broj spavaćih soba": extract_field(soup, "Spavaćih Soba", as_int=True),
            "broj kupatila": extract_field(soup, "Kupatila", as_int=True),
            "cijena": extract_field(soup, "Cijena", as_float=True),
            "stambena površina": extract_field(soup, "Stambena Površina", as_int=True),
            "zemljište": extract_field(soup, "Zemljište", as_int=True),
            "parking mjesta": extract_field(soup, "Parking Mjesta", as_int=True),
            "od mora": extract_field(soup, "Od Mora", as_int=True),
            "novogradnja": "Novogradnja" in html,
            "klima": "Klima Uređaj" in html,
            "naslov": soup.find("h2").text.strip() if soup.find("h2") else "Nema",
            "opis": extract_field(soup, "Opis", regex=True),
            "web stranica": url,
            "oglasio": extract_field(soup, "Oglasio", regex=True),
            "mobilni": re.search(r'\+382\d+', html).group(0) if re.search(r'\+382\d+', html) else "Nema",
            "broj/id oglasa": int(re.search(r"Oglas Broj.?(\d+)", html).group(1)) if re.search(r"Oglas Broj.?(\d+)", html) else "Nema",
            "zadnja promjena": re.search(r"Zadnja Promjena.?[:\s]+(.+?)<", html).group(1).strip() if re.search(r"Zadnja Promjena.?[:\s]+(.+?)<", html) else "Nema",
            "slike": [img['href'] for img in soup.select('#rea_blueimp a[href]')] or []
        }

        output_data.append(data)
        print(f"{url} scraped.")
    except Exception as e:
        print(f"Failed on {url}: {e}")

driver.quit()

import csv
with open("realitica_scraped.csv", "w", newline="", encoding="utf-8") as csvfile:
    fieldnames = list(output_data[0].keys())
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for row in output_data:
        writer.writerow(row)

print("gg")
```
